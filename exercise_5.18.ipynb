```python
#a

def f(x):
    return x**4-2*x+1

# parameters
a = 0
b = 2
N = 10
h = (b-a)/N

# derivate
def central_difference(f,x,d=1e-15):
    return (f(x+d)-f(x-d)) / (2*d)


# Trapezoidal rule
s = 0.5*f(a) + 0.5*f(b)
for k in range(1,N):
    s += f(a+k*h)
trapezoid_result = h*s



# derivate calculation
df_a = central_difference(f,a)
df_b = central_difference(f,b)


# calculate correction term
correnction_term = (1/12) * (h**2) * (df_a-df_b)

# Euler_Maclaurin
euler_maclaurin_result = trapezoid_result + correnction_term

# results
print('trapezoid result',trapezoid_result)
print('euler_maclaurin_result',euler_maclaurin_result)
```
