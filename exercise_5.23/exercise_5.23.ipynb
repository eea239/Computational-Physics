```python
# a
import numpy as np

# data
w = np.loadtxt('altitude.txt')

# derivate (It automatically and rapidly calculates derivatives using Central Differences in the interior of the matrix,
# and Forward/Backward Differences at the edges.)
h = 30000
dw_dy,dw_dx = np.gradient(w,h)

```

```python
# b
import numpy as np
import pylab as plt


# intensity
fi = 45
intensity = -((np.cos(fi)*dw_dx) + (np.sin(fi)*dw_dy)) / np.sqrt((dw_dx)**2 + (dw_dy)**2 + 1)

# visualization
plt.figure(figsize=(12,6))
plt.imshow(intensity,cmap='Greens',aspect='auto')
plt.show()
```
