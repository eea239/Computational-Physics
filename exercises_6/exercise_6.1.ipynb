```python
#a
import numpy as np

V_plus = 5.0

# A matrix
A = np.array([
    [4,-1,-1,-1],
    [-1,3,0,-1],
    [-1,0,3,-1],
    [-1,-1,-1,4],   
],float)


# v vector
v = np.array([V_plus,0.0,V_plus,0.0],float)

```

```python
#b
N = len(v)

# gaussian elimination
for m in range(N):
    # Divide by the diagonal element
    div = A[m,m]
    A[m,:] /= div
    v[m] /= div

    # Subtract from the lower
    for i in range(m+1,N):
        mult = A[i,m]
        A[i,:] -= mult*A[m,:]
        v[i] -= mult*v[m]

# Backsubstitution
solution = np.empty(N,float)
for m in range(N-1,-1,-1):
    solution[m] = v[m]
    for i in range(m+1,N):
        solution[m] -= A[m,i]*solution[i]

print(solution)


```
