```python
# a

import numpy as np
import math as math
import pylab as plt

def H(n,x):
    if n == 0:
        return 1
    if n == 1:
        return 2*x
    else:
        H0 = 1
        H1 = 2*x
        for k in range(1,n):
            H = 2*x*H1 - 2*k*H0
            H0 = H1
            H1 = H
        return H1
    
def psi(n,x):
    return 1/np.sqrt(2**n * math.factorial(n) * np.sqrt(np.pi)) * np.exp(-x**2/2) * H(n,x)
    
# graphic
x_value = np.linspace(-4,4,100)
for n in range(4):
    y = psi(n,x_value)
    plt.plot(x_value,y)
plt.show()
```

```python
# c
import numpy as np

def H(n,x):
    if n == 0:
        return 1
    if n == 1:
        return 2*x
    else:
        H0 = 1
        H1 = 2*x
        for k in range(1,n):
            H = 2*x*H1 - 2*k*H0
            H0 = H1
            H1 = H
        return H1

def psi(n,x):
    return 1/np.sqrt(2**n * math.factorial(n) * np.sqrt(np.pi)) * np.exp(-x**2/2) * H(n,x)

def integrand(n,z):
    x = np.tan(z)
    dx_dz = 1/(np.cos(z))**2
    return (x)**2 * (psi(n,x))**2 * dx_dz

def integral(n):
    a = -(np.pi/2)
    b = (np.pi/2)
    N = 100
    x_standard , w_standard = np.polynomial.legendre.leggauss(N)

    # mapping
    xp = 0.5 * (b-a) * x_standard + 0.5 * (b+a)
    wp = 0.5 * (b-a) * w_standard
    integral_result = sum(wp*integrand(n,xp))
    return np.sqrt(integral_result)
print(integral(5))
```
